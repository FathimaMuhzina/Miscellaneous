# 1. Missing Values

```python
df.isna().sum()
```

```python
features_with_na=[features for features in df.columns if df[features].isna().sum()>0]
len(features_with_na)
```

```python
#To find the % of missing values
```

```python
for features in features_with_na:
    print(features,np.round(df[features].isnull().mean(),2),'% missing values')
```

Since there are so many missing values, we need to find out if there is any relationship between the null values and the dependant variable[SalesPrice].
For that, we plot scatter plot.

```python
for features in features_with_na:
    data=df.copy()
    #Creating a variable which indicates 1 for null value and 0 for other values
    data[features] = np.where(data[features].isnull(),1,0)
    
    data.groupby(features)['SalePrice'].median().plot.bar()
    plt.title(features)
    plt.show()
```

The inference from these graphs are there are some relationship between the null values and SalePrice. Hence the same has to be filled in with some meaningful values.

# 2.Check out all the Numerical Variables

```python
numerical_features =[feature for feature in df.columns if df[feature].dtypes != 'O']
numerical_features
```

```python
len(numerical_features)
```

```python
df[numerical_features].head()
```

```python
#Separate all year columns
```

```python
year_features=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_features
```

```python
for feature in year_features:
    print(feature,df[feature].unique())
```

```python
for feature in year_features:
    print(feature,len(df[feature].unique()))
```

```python
#Checking for relationship between YrSold and Dependant feature
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price Vs Year Sold')
```

```python
#Check the relationship of all year data with SalePrice

for feature in year_features:
    if feature!='YrSold':
        data = df.copy()
        
        data[feature] = data['YrSold']- data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#Checking for discrete values
discrete_features=[feature for feature in numerical_features if len(df[feature].unique())<25 and feature not in year_features+['Id']]
len(discrete_features)
```

```python
#Find the relationship between discrete features and SalePrice
for features in discrete_features:
    data=df.copy()
    df.groupby(features)['SalePrice'].median().plot.bar()
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.title(features)
    plt.show()
```

So we have concluded that there is a reationship between the discrete numerical features and SalePrice

# 3.Distribution of Numerical Variables

```python
#Checking Continuous numerical features
cont_features=[features for features in numerical_features if features not in year_features+discrete_features+['Id']]
len(cont_features)
```

```python
for features in cont_features:
    data =df.copy()
    data[features].hist(bins=25)
    plt.xlabel(features)
    plt.ylabel('SalePrice')
    plt.title(features)
    plt.show()
```
